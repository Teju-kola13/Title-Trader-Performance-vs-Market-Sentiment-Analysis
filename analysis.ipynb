```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# LOAD DATASETS
# =========================

fear_greed = pd.read_csv("fear_greed_index.csv")
historical = pd.read_csv("historical_data.csv")

print("Fear & Greed Columns:")
print(fear_greed.columns)

print("\nHistorical Columns:")
print(historical.columns)

# =========================
# DATE CONVERSION
# =========================

fear_greed["date"] = pd.to_datetime(fear_greed["date"])

historical["Timestamp IST"] = pd.to_datetime(
    historical["Timestamp IST"],
    errors="coerce"
)

historical["date"] = historical["Timestamp IST"].dt.date
historical["date"] = pd.to_datetime(historical["date"])

# =========================
# MERGE DATASETS
# =========================

merged = pd.merge(
    historical,
    fear_greed,
    on="date",
    how="left"
)

print("\nMerged Shape:")
print(merged.shape)

print("\nMerged Sample:")
print(merged.head())

# =========================
# TOTAL PNL BY SENTIMENT
# =========================

pnl_by_sentiment = merged.groupby(
    "classification"
)["Closed PnL"].sum()

print("\nTotal PnL By Sentiment")
print(pnl_by_sentiment)

# =========================
# AVERAGE PNL
# =========================

avg_pnl = merged.groupby(
    "classification"
)["Closed PnL"].mean()

print("\nAverage PnL")
print(avg_pnl)

# =========================
# WIN RATE
# =========================

merged["Win"] = merged["Closed PnL"] > 0

win_rate = merged.groupby(
    "classification"
)["Win"].mean() * 100

print("\nWin Rate (%)")
print(win_rate)

# =========================
# TRADING VOLUME
# =========================

volume = merged.groupby(
    "classification"
)["Size USD"].sum()

print("\nTrading Volume")
print(volume)

# =========================
# SAVE OUTPUT
# =========================

merged.to_csv(
    "merged_analysis.csv",
    index=False
)

print("\nmerged_analysis.csv saved successfully")

# =========================
# CHART 1
# =========================

plt.figure(figsize=(8,5))
pnl_by_sentiment.plot(kind="bar")
plt.title("Total PnL by Sentiment")
plt.ylabel("PnL")
plt.tight_layout()
plt.savefig("pnl_by_sentiment.png")
plt.show()

# =========================
# CHART 2
# =========================

plt.figure(figsize=(8,5))
avg_pnl.plot(kind="bar")
plt.title("Average PnL by Sentiment")
plt.ylabel("Average PnL")
plt.tight_layout()
plt.savefig("average_pnl.png")
plt.show()

# =========================
# CHART 3
# =========================

plt.figure(figsize=(8,5))
win_rate.plot(kind="bar")
plt.title("Win Rate by Sentiment")
plt.ylabel("Win Rate %")
plt.tight_layout()
plt.savefig("win_rate.png")
plt.show()

# =========================
# CHART 4
# =========================

plt.figure(figsize=(8,5))
volume.plot(kind="bar")
plt.title("Trading Volume by Sentiment")
plt.ylabel("Volume")
plt.tight_layout()
plt.savefig("volume.png")
plt.show()

# =========================
# SUMMARY
# =========================

print("\n========== FINAL INSIGHTS ==========")

print("\nTotal PnL:")
print(pnl_by_sentiment)

print("\nAverage PnL:")
print(avg_pnl)

print("\nWin Rate:")
print(win_rate)

print("\nTrading Volume:")
print(volume)
```
